# Data Dictionary 

- Survived -->  1: Alive  0: Dead
- Pclass   -->  Class1 : First Class    Class2 : Second Class       Class3 : Third Class 
- SibSp  (Sibling / Spouse)   -->  0 : No has any sibling or spouse on board   Num : Has any sibling or spouse on board 
- Sex      -->  Male / Female 
- Parch  (parents or children )
- Fare : sales of ticket
- Embarked  : country that person ride from it board  ||  C : cherburg , Q :Queenstown , S: Southampton

# Sex

```python
survived  = Data[Data['Survived'] == 1 ]['Sex'].value_counts()
Dead =Data[Data['Survived'] == 0 ]['Sex'].value_counts()

df  = pd.DataFrame([survived, Dead ],index = ['Survived','Dead'])

df.plot (kind='bar') 
```

# Class

Survived Female is larger than the Survived Male 

```python
survived  = Data[Data['Survived'] == 1 ]['Pclass'].value_counts()
Dead =Data[Data['Survived'] == 0 ]['Pclass'].value_counts()


df = pd.DataFrame(data = [survived , Dead], index = ['Survived' , 'Dead'])

df.plot(kind= 'bar')
```

# Siblings and Spouses

the  number of people on class 3 that dead is grater than class1 and class2 
because the titanc ship is hitted from the position of class3  

```python
survived = Data[Data['Survived'] == 1]['SibSp'].value_counts()
Dead = Data[Data['Survived'] == 0]['SibSp'].value_counts()

df = pd.DataFrame(data= [survived , Dead] , index = ['survived','Dead'])

df.plot(kind= 'bar', stacked = True)
```

# Parents and Chidren

The number of persons who is alone has larger number of dead people 

```python
survived = Data[Data['Survived']==1]['Parch'].value_counts()
Dead = Data [Data['Survived'] == 0]['Parch'].value_counts()

df = pd.DataFrame(data= [survived,Dead], index = ['survived','Dead'])

df.plot(kind='bar')
```
